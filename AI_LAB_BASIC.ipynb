```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from google.colab import files
uploaded = files.upload()
```

```python
shifat = pd.read_csv('/content/mushroom-edibility-classification-dataset .csv')
shifat.sample(10)
```

```python
shifat = shifat.fillna(0)
shifat["class"] = shifat["class"].astype('category')
shifat["class"] = shifat["class"].cat.codes
shifat["bruises"] = shifat["bruises"].astype('category')
shifat["bruises"] = shifat["bruises"].cat.codes
shifat.sample(10)
```

```python
shifat=((shifat-shifat.min())/(shifat.max()-shifat.min()))
shifat.sample(10)
```

```python
predictors = ['cap-shape', 'cap-surface', 'cap-color', 'odor', 'stalk-shape', 'stalk-root', 'stalk-surface-above-ring', 'stalk-surface-below-ring', 'stalk-color-above-ring', 'stalk-color-below-ring', 'ring-type', 'spore-print-color', 'population', 'habitat']
X = shifat[predictors]
y = shifat['class']
print(X)
print(y)
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
# for Logistic Regression
import matplotlib.pyplot as plt
import seaborn as sns
# modle as acurracy learning
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

# for Train the model
model = LogisticRegression()
model.fit(X_train, y_train) #Training the model
predictions = model.predict(X_test)
s1 = accuracy_score(y_test, predictions)
print(s1)

#for Decision Tree
#buildin class
from sklearn.tree import DecisionTreeClassifier
clf =DecisionTreeClassifier(criterion='entropy',random_state=1)
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
s2 = accuracy_score(y_pred,y_test)
print(s2)

#for Bar Chart
plt.bar(['Logistic Regression', ' Tree'],[s1, s2])
plt.title('Accuracy Calculation')
plt.show()
```
